```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=40.6925&lon=-73.9904")
soup = BeautifulSoup(page.content, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```python
img = tonight.find("img")
desc = img['title']
print(desc)
```

```python
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
periods
```

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d["title"] for d in seven_day.select(".tombstone-container img")]
print(short_descs)
print(temps)
print(descs)
```

```python
import pandas as pd
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    "temp": temps,
    "desc":descs
})
weather
```

```python
weather.to_csv('weather.csv')
```

# Scrapping of AccuWeather

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = "https://www.accuweather.com/en/us/brooklyn/11210/daily-weather-forecast/334651"
agent = {"User-Agent":'Chrome/81.0.4044.113'}
url = requests.get(page, headers=agent)
url.status_code
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
seven_day = soup.find(class_="page-column-1")
forecast_items = seven_day.find_all(class_="content-module")
tonight = forecast_items[0]
print(tonight.prettify())
```

```python
day = tonight.find(class_="dow").get_text()
date = tonight.find(class_="sub").get_text()
high = tonight.find(class_="high").get_text()
low = tonight.find(class_="low").get_text()
phrase = tonight.find(class_="phrase").get_text()
precip = tonight.find(class_="info precip").get_text()
print(day)
print(date)
print(high)
print(low)
print(phrase)
print(precip)
```

```python
day = [sd.get_text() for sd in seven_day.select(".page-column-1 .dow")]
date = [sd.get_text() for sd in seven_day.select(".page-column-1 .sub")]
high = [sd.get_text() for sd in seven_day.select(".page-column-1 .high")]
low = [sd.get_text() for sd in seven_day.select(".page-column-1 .low")]
phrase = [sd.get_text() for sd in seven_day.select(".page-column-1 .phrase")]
precip = [sd.get_text() for sd in seven_day.select(".page-column-1 .info precip")]

print(day)
print(date)
print(high)
print(low)
print(phrase)
print(precip)



```

```python
day_tag = [sd.get_text() for sd in seven_day.select(".page-column-1 .dow")]
day_old = []
for tag in day_tag:
    days = tag.replace('\t','')
    day_old.append(days)
print(day_old)
day = []
for i in day_old:
    t = i.replace('\n','')
    day.append(t)
print(day)
```

```python
date_tag = [sd.get_text() for sd in seven_day.select(".page-column-1 .sub")]
date_old = []
for tag in date_tag:
    dates = tag.replace('\t','')
    date_old.append(dates)
print(date_old)
for b in date_old:
    c = b.replace('\n','')
    date.append(c)
print(date)
```

```python
phrase_tag = [sd.get_text() for sd in seven_day.select(".page-column-1 .phrase")]
phrase_old = []
for phr in phrase_tag:
    phrases = phr.replace('\t','')
    phrase_old.append(phrases)
print(phrase_old)
phrase = []
for k in phrase_old:
    p = k.replace('\n','')
    phrase.append(p)
print(phrase)
    
```

```python
import pandas as pd
weather = pd.DataFrame({
    "day": day,
    "date": date,
    "high": high,
    "low":low,
    "phrase":phrase
})
weather
```

```python
weather['high/low'] = weather[['high', 'low']].apply(lambda x: ''.join(x), axis=1)
```
